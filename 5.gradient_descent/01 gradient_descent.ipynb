```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1,5,100)
```

```python
y = [(xi-3)**2 -1 for xi in x]
# y = (x-3)**2 -1 
```

```python
plt.plot(x,y)
plt.show()
```

```python
def getk(x):
    return 2*(x-3)
def gety(x):
    return (x-3)**2-1
```

```python
s = np.random.randint(1,5)
s
```

```python
s = 2   # 此处应该初始化，否则运行一遍后s的值被改变了，再次跑有问题
alpha = 0.2
k = getk(s)
vals = []
vals.append(s)
#print(k)
while np.abs(k) >0.001:
    s = s - alpha * k
    k = getk(s)
    vals.append(s)
 #   print(k)
```

```python
len(vals)
```

```python
plt.plot(x,y)
plt.scatter(vals,[gety(xi) for xi in vals], color = 'r')
plt.show()
```
