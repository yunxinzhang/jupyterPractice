```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,60)
error = np.random.randn(len(x))*31-15
y = 5*x*x + 15 + error
plt.scatter(x,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures   # 增加特征
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
xtr, xt, ytr, yt = train_test_split(x.reshape(-1,1),y)
```

```python
from sklearn.linear_model import Ridge
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
def RidgeRegression(degree, alpha):
    return Pipeline([
        ("poly", PolynomialFeatures(degree=degree)),
        ("scaling", StandardScaler()),
        ("linearRegression", Ridge(alpha=alpha))
    ])
```

```python
poly_reg = RidgeRegression(3,0.0001)
poly_reg.fit(xtr,ytr)
ac = poly_reg.score(xt,yt)
print(ac)
yp = poly_reg.predict(x.reshape(-1,1))
plt.scatter(x,y)
plt.plot(x, yp, 'r')
plt.show()
```

```python
poly_reg = RidgeRegression(40,0.0001)
poly_reg.fit(xtr,ytr)
poly_reg.score(xt,yt)
yp = poly_reg.predict(x.reshape(-1,1))
plt.scatter(x,y)
plt.plot(x, yp, 'r')
plt.show()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler
def LassorRegression(degree, alpha):
    return Pipeline([
        ("poly", PolynomialFeatures(degree=degree)),
        ("scaling", StandardScaler()),
        ("lasso", Lasso(alpha=alpha))
    ])
```

```python
lasso = LassorRegression(20,0.01)
lasso.fit(xtr,ytr)
ac =lasso.score(xt,yt)
print(ac)
yp = lasso.predict(x.reshape(-1,1))
plt.scatter(x,y)
plt.plot(x, yp, 'r')
plt.show()
```
